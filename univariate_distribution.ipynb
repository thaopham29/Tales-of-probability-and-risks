# Discrete distribuion



A discrete random variable can only assume finite or countable infinite number of distinct values.

Probability distribution, or probability mass distribution of a discrete random variable X is:

\begin{align*}
 p(x) & = \mathbb{P}(X = x ) \\\
 \end{align*}

 In words, $p(x)$ is a function that assigns probabilities to each value $x$ of the random variable X, or probability that X is exact equal to some values.

 Cumulative distribution function (CDF) of a discrete random variable X is given by:

 \begin{align*}
 F_X(x) & = \mathbb{P}(X \le x ) = \sum_{t\le x} p(t) \\\
 \end{align*}

 That means, CDF of X is the probability that X less than or equal to an arbitrary value x. In case that X is a discrete random variable, CDF is the sume of all probabilities that satisfy the above condition.

 The expected value of X is:

  \begin{align*}
\mathbb{E}(X) & =  \sum_{x} x p(x) \\\
 \end{align*}

 **Example:** Consider probability function:

 \begin{align*}
X  & =\left\{
                \begin{array}{ll}
                 1  \qquad    \text{ with probability of}  \quad 0.4\\
                 2 \qquad  \ \text{ with probability of} \quad   0.25\\
                        3  \qquad  \ \text{ with probability of} \quad   0.2\\     
                        4 \qquad  \ \text{ with probability of} \quad   0.15  
                             \end{array}
              \right. \\
\end{align*}

Determine the CDF, expected value, and variance of X.


\begin{align*}
F_X(x)  & =\left\{
                \begin{array}{ll}
                 0   \quad      \quad  x<0 \\
                 0.4  \quad      \quad   1 \le x < 2\\
                  0.65  \quad  \  \quad  2 \le x < 3 \\     
                  0.85  \quad  \  \quad   3 \le x < 4 \\
                  1 \quad  \  \quad  x\ge 4
                             \end{array}
              \right. \\
\end{align*}

The expected value of X is:


  \begin{align*}
\mathbb{E}(X) & =  \sum_{x} x p(x) \\\
& = 1 \times 0.4 +  2\times 0.25 + 3  \times 0.2 + 4  \times 0.15 = 2.1
 \end{align*}

 The second moment about the origin of X is:

  \begin{align*}
\mathbb{E}(X^2) & =  \sum_{x} x^2 p(x) \\\
& = 1 \times 0.4 +  4\times 0.25 + 9  \times 0.2 + 16  \times 0.15 = 5.6
 \end{align*}

 The variance of X is

 \begin{align*}
Var(X) & =  \mathbb{E}(X^2) - (\mathbb{E}(X))^2 \\\
& = 5.6 - 2.1^2 = 1.19
 \end{align*}

 The graph below show the CDF of discrete variable X, which is the step function.

```r
#library
library(ggplot2)

# Define the discrete values of X and their probabilities
x <- c( 1,2,3,4,5)
fx<- c(0,0.4,0.25,0.2,0.15)
Fx <- cumsum(fx)
n <- length(x)

plot(x = NA, y = NA, pch = NA,
     xlim = c(0, max(x)),
     ylim = c(0, 1),
     xlab = "X ",
     ylab = "Y label",
     main = "CDF")
points(x = x[-n], y = Fx[-1], pch=19)
points(x = x[-1], y = Fx[-1], pch=1)
for(i in 1:(n-1)) points(x=x[i+0:1], y=Fx[c(i,i)+1], type="l")


```

## Binomial distribution

Let's start with a Bernoulli experiment, which composes of 2 possible outccomes: success (true) with the probability of $p$ and failure (false) with the probability of $q$. The assumption for $p$ qnd $q$ are:


*   $p+q = 1$
*   $0 < p,q<1$

Binomial experiment consist of a finite number of Bernoulli trials, let's say n. The most important assumption is that these Bernoulli trials are independent. Therefore, the outcome of a binomial experiment can be: $S_1S_2...S_n$, with:

 \begin{align*}
\left\{
                \begin{array}{9}
                 S_i = S  \qquad    \text{ with probability of}  \quad p\\  
                        S_i = F \qquad  \ \text{ with probability of} \quad   q  
                             \end{array}
              \right. \\
\end{align*}

 A discrete random variable X is said to be binomial, when X denotes the number of success in one outcome.

 In binomial distribution, we concern about the probability of $X =r $, or the probability that we can obtain r number of successes in n trials. Mathematically expressed:


  \begin{align*}
\mathbb{P}_X(x = r) & = \binom{n}{r}p^rq^{n - r}.
 \end{align*}

The formula can be decomposed into 2 components: $ \binom{n}{r} = \frac{n!}{r!(n-r)!} $, the number of ways to choose r success in n attempts; $p^r$:  the probability of getting success exactly r times in n trials; and $q^{n - r}$: the probability of getting failure in the rest trials, or (n - r) trials.


The expected value of X is:

\begin{align*}
\mathbb{E}(X) & = \sum_{x}x p(x) \\
& = \sum_{x=0}^{n} x \binom{n}{x}p^xq^{n - x} \\
& = \sum_{x=1}^{n} x. \frac{n!}{x!(n-x)!}p^xq^{n - x}\\
& = \sum_{x=1}^{n} \frac{n!}{(x- 1)!(n-x)!}!p^xq^{n - x}
 \end{align*}

 Use $u = x - 1$, we have:
\begin{align*}
\mathbb{E}(X) & = np \sum_{x=1}^{n}\frac{(n - 1)!}{(x- 1)!(n-x)!}!p^{x - 1}q^{n - x} \\
& = np \sum_{u=0}^{n -1}\frac{(n - 1)!}{(u)!(n- u -1)!}!p^{u}q^{n - u -1}
 \end{align*}

 We have $ \frac{(n - 1)!}{(u)!(n- u -1)!}!p^{u}q^{n - u -1} = \binom{n -1}{u}p^{u}q^{n - u -1} $ a binomial probability function with $(n - 1)$ trials. Therefore: $\sum_{u=0}^{n -1}\binom{n -1}{u}p^{u}q^{n - u -1} = 1$.

 Then $\mathbb{E}(X) = np$

 The variance of X is calculated with the support of

 \begin{align*}
\mathbb{E}(X(X - 1)) & = \sum_{x=0}^{n} x(x -1) \binom{n}{x}p^xq^{n - x} \\
& = \sum_{x=2}^{n}\frac{n!}{(x - 2)!(n-x)!}p^xq^{n - x}\\
& = n(n - 1)p^2\sum_{x=2}^{n} \frac{(n-2)!}{(x -2)!(n-x)!}p^{x - 2}q^{n - x}
 \end{align*}

Use $ v = x -2$, we have
\begin{align*}
\mathbb{E}(X(X - 1))  & =  n(n - 1)p^2\sum_{v = 0}^{n -2} \frac{(n-2)!}{v!(n-v -2)!}p^{v}q^{n - v -2}
 \end{align*}

 We have $\frac{(n-2)!}{v!(n-v -2)!}p^{v}q^{n - v -2} = \binom{n -2}{v}p^{v}q^{n - v- 2} $ a binomial probability funcction with $(n - 2)$ trials.

  Therefore: $\sum_{v=0}^{n -2}\binom{n -2}{v}p^{v}q^{n - v- 2} = 1$.

 Then $\mathbb{E} (X(X - 1)) = n(n - 1)p^2$

 \begin{align*}
Var(X)   & =  \mathbb{E} (X^2) - (\mathbb{E}(X))^2 \\
& = \mathbb{E} (X(X - 1)) + \mathbb{E}(X) - (\mathbb{E}(X))^2  \\
& = n(n - 1)p^2 + np - n^2p^2 \\
&= np(1-p)
 \end{align*}



 **Example**: A factory tests 100 computer chips, each of which has an independent 3% chance of being defective. Calculate:

 a) The probability that 2 chips are defective.

 b) The probability that exact 2 chips are defective given that the number of defective chips are at most 2.

Let X be a binomial random variable representing the number of defective chips, or $ X   \sim Bin (100,0.03)$.

a) $ \mathbb{P} (X =2) = \binom{100}{2}0.03^20.97^{98} = 0.225$

b)

\begin{align*}
 \mathbb{P} (X =2/X\le 2)   & = \frac{\mathbb{P} (X =2)}{\mathbb{P} (X =2) + \mathbb{P} (X =1) + \mathbb{P} (X =0)}  \\
 & = \frac{\binom{100}{2}0.03^20.97^{98}}{\binom{100}{2}0.03^20.97^{98} + \binom{100}{1}0.030.97^{99} + \binom{100}{0}0.97^{100}}\\
 & = 0.536
 \end{align*}

The graph below shows the PDF of different binomial distribution (same n, different p)

```r
# Set parameters
n <- 100
p_values <- c(0.03, 0.1, 0.5)

# Create a sequence of x values (possible outcomes)
x <- 0:n

# Plot the binomial probability distributions for the different p values
plot(x, dbinom(x, size = n, prob = p_values[1]), type = "h", col = "blue", lwd = 2,
     xlab = "x", ylab = "Probability", main = "Comparison of Binomial Distributions")
lines(x, dbinom(x, size = n, prob = p_values[2]), type = "h", col = "red", lwd = 2)
lines(x, dbinom(x, size = n, prob = p_values[3]), type = "h", col = "green", lwd = 2)

# Add a legend
legend("topright", legend = paste("p =", p_values), col = c("blue", "red", "green"), lwd = 2)

```



## Poisson distribution
Poisson distribution has many applications, especcially in actuarial science. Typically, Poisson distribution is used to model the number of random occurences of some events, which could happen in some fixed space or time. For example, in actuarial science, one often use Poisson distribution to model the number of claims per policyholder during a time period (e.g one year).

PMFs of Poission distribution is:

$ P(X =k ) = \frac{\lambda ^k}{k!}e ^{-\lambda }$, $ k = 0, 1, 2, ..., \lambda > 0$

The expected value of X is:

\begin{align*}
\mathbb{E}(X) & = \sum_{x}x p(x) \\
& = \sum_{x=0}^{\infty} x \frac{\lambda ^x}{x!}e ^{-\lambda } \\
& = \sum_{x=1}^{\infty} x \frac{\lambda ^x}{x!}e ^{-\lambda }\\
& = \sum_{x=1}^{\infty}\frac{\lambda ^x}{(x -1)!}e ^{-\lambda }
 \end{align*}

 Use $u = x -1$  then,


\begin{align*}
\mathbb{E}(X) & = \lambda\sum_{x=1}^{\infty}\frac{\lambda ^{x -1 }}{(x -1)!}e ^{-\lambda }  \\
& =  \lambda\sum_{u =0}^{\infty}\frac{\lambda ^{u}}{u!}e ^{-\lambda } \\
 \end{align*}

 Notice that $ \frac{\lambda ^{u}}{u!}e ^{-\lambda }$ is Poisson probability function. Therefore, $\sum_{u =0}^{\infty}\frac{\lambda ^{u}}{u!}e ^{-\lambda } = 1$.

 Then, $\mathbb{E}(X) = \lambda$.

 The variance of X is calculated with the support of:

 \begin{align*}
\mathbb{E}(X(X - 1)) & = \sum_{x=0}^{\infty} x(x -1)\frac{\lambda ^x}{x!}e ^{-\lambda }   \\
& = \sum_{x=2}^{\infty}x(x -1)\frac{\lambda ^x}{x!}e ^{-\lambda } \\
& = \sum_{x=2}^{\infty}\frac{\lambda ^x}{(x - 2)!}e ^{-\lambda }
 \end{align*}

 Use $v = x -2$ then:

 \begin{align*}
\mathbb{E}(X(X - 1)) & = \lambda^2\sum_{x=2}^{\infty}\frac{\lambda ^{x -2}}{(x - 2)!}e ^{-\lambda }
&= \lambda^2\sum_{v = 0}^{\infty}\frac{\lambda ^{v}}{v!}e ^{-\lambda }
 \end{align*}

 As above $\sum_{v = 0}^{\infty}\frac{\lambda ^{v}}{v!}e ^{-\lambda } = 1$, then $\mathbb{E}(X(X - 1)) = \lambda^2$.

 The variance of X is:

 \begin{align*}
Var(X)   & =  \mathbb{E} (X^2) - (\mathbb{E}(X))^2 \\
& = \mathbb{E} (X(X - 1)) + \mathbb{E}(X) - (\mathbb{E}(X))^2  \\
& = \lambda^2 + \lambda - \lambda^2 \\
&= \lambda.
 \end{align*}


**Example**: The number of claims received by an insurance company in a month is asummed to follow a Poisson distribution. The probability that the company receives less than 2 claims in a month is $8/17$ times probability that the company receives less than 3 claims in a month. Determine the probability that the company receives 5 claims in a month.

Let N the number of claims received by an insurance company in a month, or $N$

The probability that the company receives less than 2 claims means that the probability that the company receives only one claim or the company does not receive any claim during the month.

$ \mathbb{P}(N < 2) = \mathbb{P}(N = 0 ) + \mathbb{P}(N= 1) = e ^{-\lambda } + e ^{-\lambda }\lambda. $

Similarly, the probability that the company receives less than 3 claims is:

$ \mathbb{P}(N < 3) = \mathbb{P}(N = 0) + \mathbb{P}(N= 1) + \mathbb{P}(N = 2) = e ^{-\lambda } + e ^{-\lambda }\lambda + e ^{-\lambda }\frac{\lambda ^2}{2}.$

As $ \mathbb{P}(N < 2) = \frac{8}{17}\mathbb{P}(N < 3): $

\begin{align*}
 e ^{-\lambda } + e ^{-\lambda }\lambda & =\frac{8}{17}(e ^{-\lambda } + e ^{-\lambda }\lambda + e ^{-\lambda }\frac{\lambda ^2}{2})\\
 1 + \lambda & = \frac{8}{17}(1 + \lambda  + \frac{\lambda ^2}{2})\\
17 + 17\lambda & = 8 + 8\lambda + 4\lambda ^2\\
4\lambda ^2 - 9\lambda - 9 & = 0
 \end{align*}

$ \Rightarrow \lambda = 3 $ or $ \lambda = - 0.75 $ (rejected)

Hence, the probability that the company receives at least 5  claims in a month is:

$ \mathbb{P}(N = 5) = \frac{3 ^5}{5!}e ^{-3} = 0.1008. $

The code and graph below shows the PDF for Poisson distribution with different values for $\lambda$.


```r
# Define Poisson parameters
lambda_values <- c(3, 10, 50)
x_values <- 0:100 # Range of x values

# Create an empty plot
par(mfrow=c(1,1)) # Single plot
plot(NULL, xlim=c(0, 100), ylim=c(0, 0.2), xlab="X values", ylab="Probability",
     main="Poisson Distribution (PDF) Comparison")
grid(nx=NA, ny=NULL, col="gray", lty="dotted")

# Define colors
colors <- c("blue", "red", "green")

# Plot PDFs for each lambda
for (i in 1:length(lambda_values)) {
  probabilities <- dpois(x_values, lambda_values[i])
  lines(x_values, probabilities, type="h", col=colors[i], lwd=2)
}

# Add legend
legend("topright", legend=paste("Lambda =", lambda_values), col=colors, lwd=2)
```

## Negative Binomial distribution
Consider again a series of independent Bernoulli trials  with the probability of $p$ for success (true)and with the probability of $q = 1 - p$ for failure (false). In this case, we interest in the number of trials $n$ that we will get the $r^{th}$ successes. Then the discrete random variable X that represents such events is called Negative Binomial variable, or $ X  \sim Neg (r,p)$.

We have:

\begin{align*}
\mathbb{P}_X(x) & = \binom{x - 1}{r - 1}p^rq^{x - r}.
 \end{align*}

 The explanation for this formula is as followed: in order to get  the $r^{th}$ successes in $n$ trials, we need first to get $(r - 1)^{th}$ succcess in $(x -1)$ trials, then in the trial x, another success is obtained. This can be represented by the intersection of 2 events:

 M = {the first $(x - 1)$ trials have $(r - 1)$ successes}

 N = {trial x also success}

 \begin{align*}
\mathbb{P}(N) & = p \\
\mathbb{P}(M) & = \binom{x - 1}{r - 1}p^{r - 1}q^{x - r}
 \end{align*}

 Because of the independence between trials, we can have:

 \begin{align*}
 \mathbb{P}_X(x) & =\mathbb{P}(N) \times \mathbb{P}(M)  \\
 & = \binom{x - 1}{r - 1}p^rq^{x - r}.
 \end{align*}

 The negative binomial distribution can be viewed in a different way. Let Y is the number of failures before $r^{th}$ successes.

 \begin{align*}
\mathbb{P}_Y(y) & = \mathbb{P}_X(y +r) c
& = \binom{y + r - 1}{y}p^rq^y.
 \end{align*}

The expected value of X and Y:

\begin{align*}
\mathbb{E}(X) & = \frac{r}{p}\\
\mathbb{E}(Y) & = \frac{rq}{p}
 \end{align*}

 The variance of X and Y:

\begin{align*}
Var(X) & = Var(Y) = \frac{rq}{p^2}\\
\end{align*}

The graph below shows the PDF for a Poisson distribtion (which have mean equals variance) and 2 Negative Binomial  distributions (one with lower variance and one with higher variance than the Poisson distribution). It is noticeable that the Negative Binomial  distribution produces heavier tails, which makes it more suitable for modeling overdispersed discrete values (or in actuarial science, claim frequencies).







```r
x <- 0:30

# Poisson
lambda <- 10
pois <- dpois(x, lambda)

# Negative binomial with same mean
# mean = r(1-p)/p

r1 <- 20
p1 <- r1 / (r1 + lambda)   # same mean

r2 <- 5
p2 <- r2 / (r2 + lambda)   # same mean

nb1 <- dnbinom(x, size = r1, prob = p1)
nb2 <- dnbinom(x, size = r2, prob = p2)
plot(x, pois,
     type = "b",
     pch = 16,
     ylim = c(0, max(pois, nb1, nb2)),
     xlab = "Number of claims",
     ylab = "Probability",
     main = "Poisson vs Negative Binomial (same mean, different variance)")

lines(x, nb1, type = "b", pch = 17, col = 2)
lines(x, nb2, type = "b", pch = 15, col = 4)

legend("topright",
       legend = c("Poisson",
                  "NegBin (low var)",
                  "NegBin (high var)"),
       col = c(1,2,4),
       pch = c(16,17,15),
       lwd = 2)
```

# Continous distribution
A continuous random variable can take value within an interval. There are 2 functions that are very important to a continuous random variable.

Cumulative distribution function (CDF) of a continuous random variable X is denoted as $ F_X(x)$, with:
\begin{align*}
 F_X(x)  = \frac{dF_X(x)}{dx} = F'_X(x), \quad - \infty < x < \infty  \\\
 \end{align*}

That means, the CDF of X is equal to probability of X less than or equal to an arbitrary value x.

By taking the derivative of CDF of X, we got the probability density function (PDF) of X, whereever the derivative exists.

\begin{align*}
 f_X(x)  = \mathbb{P}(X\le x) \\\
 \end{align*}

 Hence,
\begin{align*}
 F_X(x) = \int_{-\infty }^{x} f(v) dv. \\\
 \end{align*}

 The expected value of X is:
\begin{align*}
 \mathbb{E}(X) & = \int_{- \infty}^{ \infty} xf(x) \,dx  \\\
 \end{align*}

**Example**: x has the density function:

 $
f_X(x) = \begin{cases}
         \frac{1}{3}x^2 , & \quad 0 \le x \le 2 \\
      0, & \text{otherwise}
         \end{cases}
$

Determine the CDF, expected value and variance of X.

The CDF is obtained by integrating the PDF:

\begin{align*}
 F_X(x) & = \int_{0}^{x} \frac{1}{3}y^2  dy  \\\
 & =  \frac{y^3}{9} \Big|_0^x = \frac{x^3}{9}
 \end{align*}

 Therefore CDF of continuous random variable X is:

  $
F_X(x) = \begin{cases}
             0,  & \quad   x < 0     \\
        \frac{x^3}{9}  , & \quad 0 \le x \le 2 \\
      1, & \quad x > 3
         \end{cases}
$

The expected value of X is:

\begin{align*}
 \mathbb{E}(X) & = \int_{0}^{2} xf(x) \,dx \ = \int_{0}^{3} x.\frac{1}{3}x^2 \,dx \\\
  & = \frac{x^4}{12}\Big|_0^2 = \frac{4}{3}.
 \end{align*}

The second moment about the origin of X:

\begin{align*}
 \mathbb{E}(X^2) & = \int_{0}^{2} x^2f(x) \,dx \ = \int_{0}^{3} x^2.\frac{1}{3}x^2,dx \\\
  & = \frac{x^5}{15}\Big|_0^2 = \frac{32}{15}.
 \end{align*}

 The variance of X is then:

\begin{align*}
Var (X)& = \mathbb{E}(X^2) - (\mathbb{E}(X))^2 \\
 & = \frac{32}{15} -  \frac{4^2}{3^2} = \frac{2}{15}.
 \end{align*}




## Uniform distribution
Let's start with the simplest form of continuous distribution. A continuous uniform random variable X has the following PDF:

$
f(x) = \begin{cases}
            \frac{1}{b-a}, & \text{if} \quad a \le x \le b \\
      0, & \text{otherwise}
         \end{cases}
$

And the CDF given by:

$
T(x) = \begin{cases}
            0 & \text{if} \quad  x \le a \\
      \frac{x - a}{b-a} & \text{if} \quad a \le x \le b \\
      1 & \text{if}\quad x \ge b
         \end{cases}
$

The expected value of X is:

\begin{align*}
 \mathbb{E}(X) & = \int_{a}^{b} xf(x) \,dx \ = \int_{a}^{b} \frac{x}{b-a} \,dx \\\
  & = \frac{x^2}{2(b-a)}\Big|_a^b = \frac{b^2  - a^2}{2(b-a)}\\
 & = \frac{a+b}{2}
 \end{align*}

In order to calculate the variance of X, we need first to determine the second moment about the origin of X, as:

\begin{align*}
 \mathbb{E}(X^2) & = \int_{a}^{b} x^2f(x) \,dx \ =\frac{x^3}{3(b-a)}\Big|_a^b \\
  & =  \frac{b^3  - a^3}{3(b-a)}= \frac{a^2 + b^2 +ab}{3}
 \end{align*}

The Variance of X is then:

$Var(X) = \mathbb{E}(X^2) - (\mathbb{E}(X))^2 =\frac{a^2 + b^2 +ab}{3} - \frac{(a+b)^2}{4} = \frac{(b - a)^2}{12} $

**Example**: The future lifetime of a machine follows a Uniform(0,10) distribution. Given that the machine has already operated for 5 years without failure, determine the probability that it will fail within the next 2 years (i.e., between years 5 and 7).

Let T the future lifetime of the machine. We have $T  \sim \rm{Uni}(0,10)$.

\begin{align*}
 \mathbb{P}(5<T<7|T>5) &  =\frac{\mathbb{P}(5<T<7)}{\mathbb{P}(T>5)} \\
  & =  \frac{F_T(7) - F_T(5)}{1 - F_T(5)}
 \end{align*}

 Given that $F_T(t)$ the CDF of random variable T with formula:

 $F_T(t) = \frac{t}{10}, \quad  0 < x < 10$

 Therefore,

  $F_T(7) = \frac{7}{10}, F_T(5) = \frac{1}{2} $

 $\frac{F_T(7) - F_T(5)}{1 - F_T(5)} = \frac{2}{5}$

 The graph below illustrates the PDF and CDF of continuous uniform distributions $T  \sim \rm{Uni}(0,10)$.


```r
# Illustration: PDF and CDF for a continuous uniform RV in the interval (0,10)
# Define the range for X
x <- seq(0,10,length.out = 100)

#Compute pdf and cdf
pdf_x <-dunif(x, min =0, max = 10)
cdf_x <-punif(x, min =0, max = 10)

#Set layout to show 2 plots side by side
par(mfrow=c(1,2))

#Plot PDF
plot(x,pdf_x,type="l",col="blue",lwd=2,
     main="PDF of Uniform(0,10)",
     xlab ="x", ylab="Density")
grid()
segments(0,0.10,0,0, col ="black",lwd =2,lty =2)
segments(10,0.10,10,0, col ="black",lwd =2,lty =2)

#Plot CDF
plot(x,cdf_x,type ="l",col ="red",lwd =2,
     main = "CDF of Uniform(0,10)",
     xlab ="x", ylab="Cumulative Probability")
grid()
lines(c(-2,0),c(0,0), col="red",lwd =2)
lines(c(10,12),c(1,1), col="red",lwd =2)

#Reset plot layout
par(mfrow = c(1,1))
```

## Exponential distribution
The next common continuous distribution is exponential distribution, which is used to measure waiting time or machine failure time.

The PDF of an exponential random variable X is:

$
f(x) = \begin{cases}
            \lambda e^{-\lambda x} & \text{if} \quad  x 	\ge 0 \\
      0 & \text{otherwise}
         \end{cases}
$

The parameter $\lambda$ is called rate parameter. The CDF is then:

\begin{align*}
F_X(x) &  = \mathbb{P}(X \le x) = \int_{0}^{x} \lambda e^{-\lambda v}  \,dv\\
  & =   - e^{-\lambda v}\Big|_0^x = 1 -  e^{-\lambda x}
 \end{align*}

The expected value can be determined with support of intergration by parts:

\begin{align*}
 \mathbb{E}(X)  & = \int_{0}^{ \infty} x \lambda e^{-\lambda x}  \,dx\\
  & =   \biggl( - x e^{-\lambda x} \biggl)\Big|_0^\infty + \int_{0}^{ \infty} e^{-\lambda x}\, dx \\
  & = \biggl( \frac{-1}{\lambda}e^{-\lambda x} \biggl)\Big|_0^\infty = \frac{1}{\lambda}
 \end{align*}

 Again using integration by parts twice, we can obtain the second moment about the origin of X twice:

 \begin{align*}
 \mathbb{E}(X^2)  & = \int_{0}^{ \infty} x^2 \lambda e^{-\lambda x}  \,dx\\
  & =   \biggl( - x^2 e^{-\lambda x} \biggl)\Big|_0^\infty + \int_{0}^{ \infty} 2xe^{-\lambda x}\, dx \\
  & = 2\biggl( - \frac{x}{\lambda} e^{-\lambda x} \biggl)\Big|_0^\infty + \frac{2}{\lambda}\int_{0}^{ \infty}e^{-\lambda x} \,dx\\
  & = - \frac{2}{\lambda^2}\biggl(  e^{-\lambda x} \biggl)\Big|_0^\infty = \frac{2}{\lambda^2}.
 \end{align*}

There the variance is:

$Var(X) = \mathbb{E}(X^2) - (\mathbb{E}(X))^2 = \frac{2}{\lambda^2} - \frac{1}{\lambda^2} = \frac{1}{\lambda^2}.$

In some cases, the scale parameter $\beta$ is used in stead of  $\lambda $,
 with $\lambda =\frac{1}{\beta}$. The PDF then beccomes:

 $ f(x) = \begin{cases}
            \frac{1}{\beta} e^{-\frac{1}{\beta} x} & \text{if} \quad  x 	\ge 0 \\
      0 & \text{otherwise}
         \end{cases}
 $

 The interesting property of exponential distribution is memoryless, mathematically we have:

  \begin{align*}
 \mathbb{P}(X > u + t| X > t) &  = \mathbb{P}(X > u), & \text{for} \quad t, u \ge 0
  \end{align*}

With, $\mathbb{P}(X > u) = 1 -  \mathbb{P}(X \le u) = 1 - F_X(u) = e^{-\lambda u} $

 \begin{align*}
 \mathbb{P}(X > u + t| X > t) & = \frac{\mathbb{P}\Bigl(X > u + t ~ \text{and}~ X > t\Bigl) }{\mathbb{P}(X >  t)}\\
  & =   \frac{\mathbb{P}\Bigl(X > u + t) }{\mathbb{P}(X >  t)}  = \frac{e^{-\lambda(u+t)} }{e^{-\lambda u} }\\
  & = e^{-\lambda u}= \mathbb{P}(X > u).
 \end{align*}

**Example**: Time to repair a machine (in hours) x is exponential distributed with parameter $\lambda = 0.4$. Determine

*  The CDF of $X$
*  $\mathbb{P}(X>3)$?
*  $\mathbb{P}(X>6| X >4)$?

The CDF of $X$ is: $F_X(x)  =  1 - e^{-0.4 x}$


$\mathbb{P}(X>3) = 1 - \mathbb{P}(X \le 3) = 1  - e^{-0.4\times 3}  = 0.6988$

 $\mathbb{P}(X>6| X >4) = \mathbb{P}(X>(4+2)| X >4) = \mathbb{P}(X>2) = 1  - e^{-0.4\times 2}  = 0.55067$










```r
# Set lambda
lambda <- 0.4

# Define x (range from 0 to 15)
x <- seq(0,15,length.out = 300)

#Compute PDF and CDF
pdf_x <- dexp(x,rate = lambda)
cdf_x <- pexp(x,rate = lambda)

#Set layout to show plots side by side
par(mfrow =c(1,2))

#Plot PDF
plot(x,pdf_x, type ="l",col="blue",lwd=2,
     main ="PDF of Exponential(0,4)",
     xlab="x", ylab ="Density")
grid()

#Plot CIDF
plot(x,cdf_x, type ="l",col="red",lwd=2,
     main ="CCDF of Exponential(0,4)",
     xlab="x", ylab ="Cumulative Probability")
grid()

#Reset layout
par(mfrow =c(1,1))


```

## Normal distribution

A continuous random variable X obeying Normal distribution, with parameter $\mu \in \mathbb{R}$ and $\sigma > 0$ has the PDF:

\begin{align*}
f_X(x) &  = \frac{1}{\sqrt{2\pi}}exp\biggl(-\frac{1}{2\sigma^2}(x - \mu)^2\biggl)
  \end{align*}

  Then X will have:

  $\mathbb{E}(X) = \mu \quad  $ and  $\quad  Var(X) = \sigma^2$.

 Calculation for normal distribution is carried out with the support of a standard normal random variable $Z = \frac{ X -\mu }{\sigma}$ .

 **Example**: Let X follow a normal distribution with mean 10, and variance 4.
*    $\mathbb{P}(X>14)$?

\begin{align*}
 \mathbb{P}(X>14) &  = \mathbb{P}(\frac{ X -10 }{2} > \frac{ 14 -10 }{2}) \\
 & =  \mathbb{P}(Z>2) \\
 & = 1 - \mathbb{P}(Z  \leq 2) \\
 & = 1 - \Phi(2) = 1 - 0.9772 = 0.0228
  \end{align*}
*   $S(X) = 0.05$ where $S(X)$ survival function of X.
  \begin{align*}
 S(X) & =  \mathbb{P}(X>x) = 0.05\\
\Leftrightarrow \mathbb{P}(X  \leq x)  & = 1 - 0.05 = 0.95 \\
& = \mathbb{P}(\frac{ X -10 }{2} < \frac{ x -10 }{2}) \\
& = \mathbb{P}( Z <\frac{ x -10 }{2}) = 0.95
 \end{align*}

 Therefore,

  \begin{align*}
 \frac{ x -10 }{2} & =  1.65\\
 \Rightarrow x & = 13.3
\end{align*}



The graph below shows the comparison of normal distribution with same mean (100) and different standard deviations.

```r
# Define parameters
mu <- 100  # Expected value
sigma_values <- c(30, 50, 100)  # Standard deviations
colors <- c("red", "blue", "green")  # Colors for different variances


# Create sequence for x-axis
x <- seq(0, 200, length=1000)

# Plot the normal distributions
plot(x, dnorm(x, mean=mu, sd=sigma_values[1]), type="l", col=colors[1], lwd=2,
     ylim=c(0, 0.015), xlab="x", ylab="Density",
     main="Normal Distributions with Different Variances")

# Add other distributions
for (i in 2:length(sigma_values)) {
  lines(x, dnorm(x, mean=mu, sd=sigma_values[i]), col=colors[i], lwd=2)
}

# Add legend
legend("topright", legend=paste("Variance =", sigma_values^2),
       col=colors, lwd=2)
```

## Gamma distribution

The Gamma function is defined by:

\begin{align*}
\Gamma(\alpha) = \int_{0}^{\infty} x^{\alpha - 1}e^{-x}dx
  \end{align*}

  For example,

\begin{align*}
\Gamma(1) = \int_{0}^{\infty} e^{-x}dx = -e^{-x}\Big|_0^\infty = 1
  \end{align*}

For $\alpha > 1$, using integration by part we have,

\begin{align*}
\Gamma(\alpha) & = -x^{\alpha - 1}e^{-x}\Big|_0^\infty  + \int_{0}^{\infty} (\alpha -1)x^{\alpha - 2}e^{-x}dx\\
&= (\alpha -1)\int_{0}^{\infty}x^{\alpha - 2}e^{-x}dx\\
& = (\alpha -1)\Gamma(\alpha -1)
  \end{align*}

With $\alpha > 1$ and as an integer:
\begin{align*}
\Gamma(\alpha) & =  (\alpha -1)\Gamma(\alpha -1)\\
&= (\alpha -2)(\alpha -1)\Gamma(\alpha -2)\\
&.\\
&.\\
&.\\
& = (\alpha -1)(\alpha -2)...3.2.1 = (\alpha -1)!
  \end{align*}

A Gamma random variable X has the PDF as followed:

$
f(x) = \begin{cases}
            \frac{x^{\alpha - 1}\tau^\alpha e^{(-x\tau)}}{\Gamma(\alpha)}, & \text{if} \quad  x \ge 0 \\
      0, & \text{otherwise}
         \end{cases}
$

The expected value of X is:

\begin{align*}
 \mathbb{E}(X) & = \int_{0}^{\infty} xf(x) \,dx \ \\
  & = \int_{0}^{\infty} x\frac{x^{\alpha - 1}\tau^\alpha e^{(-x\tau)}}{\Gamma(\alpha)} dx \\
 & = \frac{1}{\tau\Gamma(\alpha)}\int_{0}^{\infty}\tau (x\tau)^{\alpha} e^{(-x\tau)}dx\\
 & = \frac{\Gamma(\alpha + 1)}{\tau\Gamma(\alpha)} = \frac{\alpha}{\tau}
 \end{align*}

 The second moment about the origin of X is:

 \begin{align*}
 \mathbb{E}(X^2) & = \int_{0}^{\infty} x^2f(x) \,dx \ \\
  & = \int_{0}^{\infty} x^2\frac{x^{\alpha - 1}\tau^\alpha e^{(-x\tau)}}{\Gamma(\alpha)} dx \\
 & = \frac{1}{\tau\Gamma(\alpha)}\int_{0}^{\infty}x^{(\alpha+1)}\tau ^{\alpha} e^{(-x\tau)}dx\\
 & = \frac{\Gamma(\alpha +2)}{\tau^2\Gamma(\alpha)} \int_{0}^{\infty}\frac{x^{(\alpha+1)}\tau ^{(\alpha + 2)} e^{(-x\tau)}dx}{\Gamma(\alpha +2)}\\
 & = \frac{\Gamma(\alpha +2)}{\tau^2\Gamma(\alpha)}\\
 & =  \frac{(\alpha +1)\Gamma(\alpha +1)} \\
 = \frac{(\alpha +1)\alpha }{\tau^2}
 \end{align*}

The variance is:

$Var(X) = \mathbb{E}(X^2) - (\mathbb{E}(X))^2 = \frac{(\alpha +1)\alpha }{\tau^2} - \frac{\alpha^2 }{\tau^2} = \frac{\alpha }{\tau^2}$



```r
x <- seq(0, 40, length = 500)

# parameters (same mean = a/b = 10)

a1 <- 1; b1 <- 0.1
a2 <- 2; b2 <- 0.2
a3 <- 5; b3 <- 0.5

pdf1 <- dgamma(x, shape = a1, rate = b1)
pdf2 <- dgamma(x, shape = a2, rate = b2)
pdf3 <- dgamma(x, shape = a3, rate = b3)

cdf1 <- pgamma(x, shape = a1, rate = b1)
cdf2 <- pgamma(x, shape = a2, rate = b2)
cdf3 <- pgamma(x, shape = a3, rate = b3)

par(mfrow = c(1,2))

# ----- PDF -----
plot(x, pdf1, type = "l", lwd = 2,
     ylim = c(0, max(pdf1,pdf2,pdf3)),
     xlab = "x", ylab = "Density",
     main = " Gamma PDF")

lines(x, pdf2, col = 2, lwd = 2)
lines(x, pdf3, col = 4, lwd = 2)

legend("topright",
       legend = c("a=1, b=0.1",
                  "a=2, b=0.2",
                  "a=5, b=0.5"),
       col = c(1,2,4),
       lwd = 2)

# ----- CDF -----
plot(x, cdf1, type = "l", lwd = 2,
     ylim = c(0,1),
     xlab = "x", ylab = "CDF",
     main = "Gamma CDF ")

lines(x, cdf2, col = 2, lwd = 2)
lines(x, cdf3, col = 4, lwd = 2)

legend("bottomright",
       legend = c("a=1, b=0.1",
                  "a=2, b=0.2",
                  "a=5, b=0.5"),
       col = c(1,2,4),
       lwd = 2)
```
