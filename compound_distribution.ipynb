## Brief overview

Compound distribution function of a sum that plays an important part in actuarial science is  briefly introduced in this section, which was mainly based on... The compound sum distribution is considered as one special case of mixture distribution, in which the random variable X is distributed with a distribution function depending on an unknown parameter $\theta$, which is also considered as a random variable. The resulted distribution function of X and  $\theta$ is described as the following:

\begin{align*}
F_{X_\Theta} (x)= \int_{0}^{\infty} F_{X_\theta}(x) d F_{\Theta}(\theta)
\end{align*}

General formula for a compound sum is:
\begin{align}
\tag{1}
 S = \sum_{i=1}^{N} C_i
\end{align}


 with 2 assumptions: (i)  N is  a discrete random variable and independent from $C_1$,$C_2$,...; and  (ii) the random variables  $C_1$,$C_2$,...  are  assumed to be independent and identically distributed (iid).


The most well-known compound distributions are: compound Poisson distribution, compound Binomial distribution, and compound Negative Binomial distribution.


## Moment
Asumme S a compound distribution as in (1).  Let X be a random variable distributed as $X_i$.

<ol type ="i">
<li> Expected value (first moment):</li>

We can obtain the expected value of compound sum S using conditional expectation.

\begin{align*}
\mathbb {E}[S] &= {\mathbb {E}}  [\mathbb {E} [S/N]]\\
& =  \sum_{i =1}^{\infty} \mathbb {E} [X_1 + ... X_N / N = n] \mathbb {P}[N = n]\\
& = \sum_{i =1}^{\infty} \mathbb {E} [X_1 + ... X_n / N = n] \mathbb {P}[N = n]\\
& = \sum_{i =1}^{\infty} \mathbb {E} [X_1 + ... X_n ] \mathbb {P}[N = n]\\
& = \sum_{i =1}^{\infty} n \mathbb {E} [X]\mathbb {P}[N = n]\\
& = \mathbb {E} [X]\mathbb {E} [N]
\end{align*}

<li> Variance </li>

\begin{align*}
Var [S]  &= \mathbb {E}  [Var[S/N]] + Var  [\mathbb {E} [S/N]] \\
& = \mathbb {E}  [N Var[X]] + Var[N\mathbb {X} ] \\
& = \mathbb {E}[N] Var[X] + (\mathbb {E}[X])^2Var[N]
\end{align*}

# Special cases of compound distribution





## Popular compound distribution

This section discusses two special cases of compound distribution, which are essential in actuarial science. Two important variables in many insurance problems are claim frequency, or the number of claim, which is represented by N in compound ditrbution and claim severity, or the amount of claim, denoted by $X_i$. We often use Poisson distribution (whose expected value and variance are equal) or Negative Binomial distribution (which allows for more  overdispersed discrete values) for N. Claim severity, on the other hand, is often modelled with Gamma distribution, log-normal distribution or even Pareto.

### Compound Poisson distribution: with $N \sim  \text{Poi} (\lambda)$.

A special characteristic of compound Poisson distribution is that sum of independent compound Poisson random variables also follows a compound Poisson distribution. Formally, S is the sum of  compound sums $S_1, S_2, . . ., S_m$, that is:

\begin{align*}
S = S_1 + S_2 + . . . + S_m
\end{align*}
where:

$S_i =  \sum_{j =1}^{N_i} X_{ij}$, $ N_i \sim \text{Poi} (\lambda_i)$ ;  $ X_{ij}$ are iid; $F_i(x) =  \mathbb {P} (X_{i1} \leq x)$ , for $i = 1, 2, . . ., m$

Therefore, S follows compound Poisson distribution with $ \lambda= \sum_{i=1}^{m} \lambda_i$ and  $F(x) =  \sum_{i =1}^{m} \frac{\lambda_i}{\lambda}F_i(x)   $.


### Compound Negative Binomial distribution

Consider a mixed Poisson distribution, in which the parameter $\lambda$ is replaced by a random variable $\Lambda$, given $ \Lambda \sim \text{Gamma} (a,b)$  with the probability density function of:

\begin{align*}
f(x) = \frac{b^{a}}{\Gamma(a)} x^{a -1}e^{-bx}
\end{align*}

We can obtain the probability mass function of N as:

\begin{align*}
\mathbb {P} [N  = k ] & = \int_{0}^{\infty} \mathbb {P} [N  = k / \Lambda =\lambda ] f(\lambda)d\lambda \\
& = \frac{\Gamma(a + k)}{k!\Gamma(a)}  \bigg(\frac{b}{b+1} \bigg)^{a} \bigg(\frac{1}{b+1} \bigg)^{k}
\end{align*}

So, $ N \sim \text{NegBin} (a,  \frac{b}{b+1})$. The compound Negative Binomial sum, therefore, is similar to the mixed compound Poisson sum, with the mixing parameter following the Gamma distribution.

## Panjer's recursion formula

For these types of compound distribution, we can use Panjer's recursion formula developed in  \cite{panjer} to calculate the $\mathbb {P} (S = s)$ in the case where the random variables $X_1$, $X_2$, . . . are discrete and non-negative.

More precisely, if the probability mass function of  N  for some values of  a and b is

\begin{align*}
p_N (n) = ( a +  \frac{b}{ n}) p_N(n - 1),     n = 1,2, ....
\end{align*}
then the probability of the sum S is:


\begin{align*}
p_S(0)  & =\left\{
                \begin{array}{ll}
                 \mathbb {P} (N = 0),   \qquad    \text{ if} \quad  p_X(0) = 0 \\
                 m_N(\text{ln} p_X (0)),  \qquad  \text{ if}  \quad p_X(0) > 0\\
                             \end{array}
              \right. \\
p_S(s )  & =  \frac{1}{ 1 - a. p_X(0)}  \sum_{i=1}^{s} (a+ \frac{bi}{ s})p_X(i)p_S(s - i)  \qquad  s =1,2 . . .
 \end{align*}
 where $m_N(\text{ln} p_X (0))$ is the moment generating function of N at $\text{ln} p_X (0)$.

 Compound Poisson distribution for example, $a = 0$ and $b = \lambda \geq 0 $.

\begin{align}
p_S(0)  & = e^{ - \lambda (1 - p_X(0) )} \nonumber\\
p_S(s )  & =  \frac{1}{ s}  \sum_{i=1}^{s}i \lambda   p_X(i)p_S(s - i)
\end{align}


## Some examples

**Example 2**: Compound Poisson distribution with simple, discrete claim severity.

Let X follow compound Poisson distribution, $X =\sum_{i =1}^{N} C_i  $, in which $ N  \sim \text{Poisson} (0.1)$, and $C_i$ are iid as C:

 \begin{align*}
C & =\left\{
                \begin{array}{ll}
                 0   \qquad    \text{ with probability of}  \quad 0.2\\
                 1  \qquad  \ \text{ with probability of} \quad   0.7\\
                        2  \qquad  \ \text{ with probability of} \quad   0.1\\       
                             \end{array}
              \right. \\
         \end{align*}

  Use Panjer's recursion formula to obtained the distribution of X.

   
  \begin{align*}
 \mathbb {P}(X = 0)  & = e^{-0.1(1-0.2)} = 0.92  \\
  \mathbb {P}(X = 1) & = 0.1 \times0.7  \times0.92 = 0.0646  \\
  \mathbb {P}(X = 2) & =   \frac{1}{2} ( 0.1 \times0.7  \times0.646 +  0.1 \times0.2  \times0.92) = 0.0115 \\
  \mathbb {P}(X = 3) & =  0.0007\\
  \mathbb {P}(X = 4) & = 0.00007 \\
    \mathbb {P}(X = 5) & = 0.0000038
  \end{align*}
  
 Alternatively, the recursive formula could be implemented using function agregateDist in R. The results are similar to those calculated earlier.



```r
install.packages('actuar')
library(actuar)
library(ggplot2)
pC1<-c(0.2,0.7,0.1)
lambda1<- 0.1

FS<-aggregateDist("recursive",model.freq ="poisson", model.sev=pC1,lambda=lambda1,tol=1e-6)

#plot(FS)
pS<-diff(FS)
xS <-knots(FS)
sum(pS)
df <-data.frame(pS,xS)

ggplot(df,aes(x=xS,y=pS)) +
  geom_bar(stat="identity", color="black",width = 0.2,size=1) +
 labs(x="x",y="probability")+
theme(axis.text=element_text(size=24,,face="bold"), axis.title=element_text(size=22),
legend.text = element_text( size = 25))
```
