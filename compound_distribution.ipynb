## Brief overview

Compound distribution function of a sum that plays an important part in actuarial science is  briefly introduced in this section, which was mainly based on \cite{assdom}. The compound sum distribution is considered as one special case of mixture distribution, in which the random variable X is distributed with a distribution function depending on an unknown parameter $\theta$ that is also considered as a random variable. The resulted distribution function of X and  $\theta$ is described as the following:

\begin{align*}
F_{X_\Theta} (x)= \int_{0}^{\infty} F_{X_\theta}(x) d F_{\Theta}(\theta)
\end{align*}

General formula for a compound sum is:
\begin{align}
\tag{1}
 S = \sum_{i=1}^{N} X_i
\end{align}


 with 2 assumptions: (i)  N is  a discrete random variable and independent from $X_1$,$X_2$,...; and  (ii) the random variables  $X_1$,$X_2$,...  are  assumed to be independent and identically distributed (iid).


The most well-known compound distributions are: compound Poisson distribution, compound Binomial distribution, and compound Negative Binomial distribution.


## Moment
Asumme S a compound distribution as in (1).  Let X be a random variable distributed as $X_i$.

<ol type ="i">
<li> Expected value (first moment):</li>

We can obtain the expected value of compound sum S using conditional expectation.

\begin{align*}
\mathbb {E}[S] &= {\mathbb {E}}  [\mathbb {E} [S/N]]\\
& =  \sum_{i =1}^{\infty} \mathbb {E} [X_1 + ... X_N / N = n] \mathbb {P}[N = n]\\
& = \sum_{i =1}^{\infty} \mathbb {E} [X_1 + ... X_n / N = n] \mathbb {P}[N = n]\\
& = \sum_{i =1}^{\infty} \mathbb {E} [X_1 + ... X_n ] \mathbb {P}[N = n]\\
& = \sum_{i =1}^{\infty} n \mathbb {E} [X]\mathbb {P}[N = n]\\
& = \mathbb {E} [X]\mathbb {E} [N]
\end{align*}

<li> Variance </li>

\begin{align*}
Var [S]  &= \mathbb {E}  [Var[S/N]] + Var  [\mathbb {E} [S/N]] \\
& = \mathbb {E}  [N Var[X]] + Var[N\mathbb {X} ] \\
& = \mathbb {E}[N] Var[X] + (\mathbb {E}[X])^2Var[N]
\end{align*}

## Popular compound distribution
### Compound Poisson distribution: with $N \sim  \text{Poi} (\lambda)$.

A special characteristic of compound Poisson distribution is that sum of independent compound Poisson random variables also follows a compound Poisson distribution. Formally, S is the sum of  compound sums $S_1, S_2, . . ., S_m$, that is:

\begin{align*}
S = S_1 + S_2 + . . . + S_m
\end{align*}
where:

$S_i =  \sum_{j =1}^{N_i} X_{ij}$, $ N_i \sim \text{Poi} (\lambda_i)$ ;  $ X_{ij}$ are iid; $F_i(x) =  \mathbb {P} (X_{i1} \leq x)$ , for $i = 1, 2, . . ., m$

Therefore, S follows compound Poisson distribution with $ \lambda= \sum_{i=1}^{m} \lambda_i$ and  $F(x) =  \sum_{i =1}^{m} \frac{\lambda_i}{\lambda}F_i(x)   $.

## Panjer's recursion formula

For these types of compound distribution, we can use Panjer's recursion formula developed in  \cite{panjer} to calculate the $\mathbb {P} (S = s)$ in the case where the random variables $X_1$, $X_2$, . . . are discrete and non-negative.

More precisely, if the probability mass function of  N  for some values of  a and b is

\begin{align*}
p_N (n) = ( a +  \frac{b}{ n}) p_N(n - 1),     n = 1,2, ....
\end{align*}
then the probability of the sum S is:


\begin{align*}
p_S(0)  & =\left\{
                \begin{array}{ll}
                 \mathbb {P} (N = 0),   \qquad    \text{ if} \quad  p_X(0) = 0 \\
                 m_N(\text{ln} p_X (0)),  \qquad  \text{ if}  \quad p_X(0) > 0\\
                             \end{array}
              \right. \\
p_S(s )  & =  \frac{1}{ 1 - a. p_X(0)}  \sum_{i=1}^{s} (a+ \frac{bi}{ s})p_X(i)p_S(s - i)  \qquad  s =1,2 . . .
 \end{align*}
 where $m_N(\text{ln} p_X (0))$ is the moment generating function of N at $\text{ln} p_X (0)$.

 Compound Poisson distribution for example, $a = 0$ and $b = \lambda \geq 0 $.

\begin{align}
p_S(0)  & = e^{ - \lambda (1 - p_X(0) )} \nonumber\\
p_S(s )  & =  \frac{1}{ s}  \sum_{i=1}^{s}i \lambda   p_X(i)p_S(s - i)
\end{align}
